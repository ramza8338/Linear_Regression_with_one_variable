```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

# Define the linear regression model and cost function
def linear_regression(x, w, b):
    return w * x + b

def compute_cost(x, y, w, b):
    m = x.shape[0]
    predictions = linear_regression(x, w, b)
    cost = np.mean((predictions - y) ** 2) / 2
    return cost

# Define the gradient descent function
def gradient_descent(x, y, w, b, alpha, num_iters):
    m = x.shape[0]
    cost_history = []
    
    for i in range(num_iters):
        predictions = linear_regression(x, w, b)
        dw = (1 / m) * np.dot(x, (predictions - y))
        db = (1 / m) * np.sum(predictions - y)
        
        w -= alpha * dw
        b -= alpha * db
        
        cost = compute_cost(x, y, w, b)
        cost_history.append(cost)
        
        if i % 100 == 0:
            print(f"Iteration {i}: Cost {cost}")

    return w, b, cost_history

# Initialize parameters and run gradient descent
w_init = 0
b_init = 0
iterations = 10000
learning_rate = 0.01

w_final, b_final, J_hist = gradient_descent(x_train, y_train, w_init, b_init, learning_rate, iterations)

print(f"Final parameters: w = {w_final}, b = {b_final}")

# Plot cost versus iteration
plt.plot(J_hist)
plt.title("Cost vs. Iteration")
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.show()

# Make predictions
x_test = np.array([1.0, 1.2, 2.0])
predictions = linear_regression(x_test, w_final, b_final)
print(f"Predictions: {predictions}")

# You can add more visualization and customization as needed.
```
